```python
from sklearn.model_selection import train_test_split
```

```python
X, y = mglearn.datasets.make_forge()
X_egitim, X_test, y_egitim, y_test=train_test_split(X,y,random_state=0)
from sklearn.neighbors import KNeighborsClassifier
snf=KNeighborsClassifier(n_neighbors=3)
snf.fit(X_egitim,y_egitim)
```

```python
snf.predict(X_test)
```

```python
snf.score(X_test,y_test)
```

## Practice

### Importing data

```python
from sklearn.datasets import load_breast_cancer
```

```python
kanser=load_breast_cancer()
```

```python
kanser.keys()
```

### Exploring the data

```python
print(kanser['DESCR'])
```

### Preprocessing the data

```python
X_egitim, X_test, y_egitim, y_test=train_test_split(kanser.data, kanser.target,
                                                   stratify=kanser.target, random_state=66)
```

```python
egitim_dogruluk=[]
test_dogruluk=[]
```

```python
komsuluk_sayisi=range(1,11)
```

```python
import matplotlib.pyplot as plt
for n_komsuluk in komsuluk_sayisi:
    snf=KNeighborsClassifier(n_neighbors=n_komsuluk)
    snf.fit(X_egitim,y_egitim)
    egitim_dogruluk.append(snf.score(X_egitim,y_egitim))
    test_dogruluk.append(snf.score(X_test, y_test))
plt.plot(komsuluk_sayisi, egitim_dogruluk, label="Eƒüitim dogruluk")
plt.plot(komsuluk_sayisi, test_dogruluk , label="Test doƒüruluk ")
plt.ylabel("Doƒüruluk")
plt.xlabel("n-kom≈üuluk")
plt.legend()
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
X,y=mglearn.datasets.make_wave(n_samples=40)
```

```python
reg.fit(X_egitim, y_egitim)
```

```python
reg.score(X_test,y_test)
```
