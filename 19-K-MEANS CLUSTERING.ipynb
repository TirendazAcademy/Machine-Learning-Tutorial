```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
X,y_gercek=make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X[:,0],X[:,1],s=50)
```

# K-means clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(X)
```

```python
y_kmeans=kmeans.predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],s=50, c=y_kmeans, cmap='viridis')
centers=kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='black',s=200,alpha=0.5)
```

## K-means is limited to linear cluster boundaries

```python
from sklearn.datasets import make_moons
```

```python
X,y=make_moons(200, noise=.05, random_state=0)
```

```python
labels=KMeans(2, random_state=0).fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],s=50, c=labels, cmap='viridis')
```

```python
from sklearn.cluster import SpectralClustering
```

```python
model=SpectralClustering(n_clusters=2, affinity='nearest_neighbors',
                         assign_labels='kmeans')
labels=model.fit_predict(X)
plt.scatter(X[:,0],X[:,1],s=50, c=labels, cmap='viridis')
```

## Clustering for big data

```python
ax=plt.axes(xticks=[],yticks=[])
ax.imshow(china)
```

```python
china.shape
```

```python
veri=veri.reshape(427*640,3)
```

```python
veri.shape
```

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
kmeans=MiniBatchKMeans(16)
```

```python
kmeans.fit(veri)
```
