```python
categories=['rec.motorcycles','rec.sport.baseball','comp.graphics','rec.sport.hockey']
```

```python
from sklearn.datasets import load_files
```

```python
twenty_train=load_files('Data/20newsbydate/20news-bydate-train/',
                       categories=categories,
                       shuffle=True,
                       random_state=42,
                       encoding='utf-8',
                       decode_error='ignore')
```

```python
type(twenty_train)
```

```python
twenty_train.target_names
```

```python
len(twenty_train.data)
```

```python
twenty_train.target[:10]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect=CountVectorizer()
```

```python
X_train_counts=count_vect.fit_transform(twenty_train.data)
```

```python
X_train_counts.shape
```

## tfâ€“idf technic

```python
from sklearn.feature_extraction.text import TfidfTransformer
tf_transformer=TfidfTransformer(use_idf=False).fit(X_train_counts)
X_train_tf=tf_transformer.transform(X_train_counts)
```

```python
X_train_tf.shape
```

## Building the model

```python
from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB().fit(X_train_tf, twenty_train.target)
```

```python
docs_new=['brake-lamp is good','this computer is fast']
```

```python
X_new_count=count_vect.transform(docs_new)
X_new_tf=tf_transformer.transform(X_new_count)
```

## Predicting the data

```python
predicted=clf.predict(X_new_tf)
```

```python
for doc, category in zip(docs_new, predicted):
    print('%r=>%s' %(doc,twenty_train.target_names[category]))
```

## Pipeline

```python
from sklearn.pipeline import Pipeline
```

```python
text_clf=Pipeline([('vect', CountVectorizer()),
                  ('tfidf', TfidfTransformer()),
                  ('clf',MultinomialNB())])
```

```python
text_clf.fit(twenty_train.data, twenty_train.target)
```

```python
twenty_test=load_files('Data/20newsbydate/20news-bydate-test/',
                       categories=categories,
                       shuffle=True,
                       random_state=42,
                       encoding='utf-8',
                       decode_error='ignore')
```

```python
docs_test=twenty_test.data
```

## Predicting the model

```python
predicted=text_clf.predict(docs_test)
```

```python
import numpy as np
np.mean(predicted==twenty_test.target)
```

## SVM

```python
from sklearn.linear_model import SGDClassifier
```

```python
text_clf=Pipeline([('vect', CountVectorizer()),
                  ('tfidf', TfidfTransformer()),
                  ('clf',SGDClassifier(loss='hinge',
                                      penalty='l2',
                                      alpha=1e-3,
                                      random_state=42,
                                      max_iter=5,
                                      tol=None))])
```

```python
text_clf.fit(twenty_train.data, twenty_train.target)
predicted=text_clf.predict(docs_test)
np.mean(predicted==twenty_test.target)
```

```python
from sklearn import metrics
metrics.confusion_matrix(twenty_test.target, predicted)
```
